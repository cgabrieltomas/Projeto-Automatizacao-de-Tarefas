```python
import pyautogui
import time
```

```python
pyautogui.PAUSE = 0.5

pyautogui.hotkey("ctrl","t")
pyautogui.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
time.sleep(2)
pyautogui.press("enter")
pyautogui.click(x=924, y=469)
pyautogui.write("meu_login")
pyautogui.press("tab")
pyautogui.write("minha_senha")
pyautogui.click(x=957, y=654)
time.sleep(2)
pyautogui.click(x=453, y=462)
pyautogui.click(x=1666, y=235)
pyautogui.click(x=1384, y=741)
time.sleep(5)
```

```python
import pyperclip
pyautogui.PAUSE= 1

pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://mail.google.com/mail/u/0/#inbox")
pyautogui.press("enter")
time.sleep(5)
pyautogui.click(x=92, y=256)
pyautogui.write("pythonimpressionador@gmail.com")
pyautogui.press("tab")
pyautogui.press("tab")
pyperclip.copy(f"Relatório de Venda") #usado pois o pyautogui.write não lê caracteres especiais
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab")
texto=f"""
Segue o relatório de compras

Total Gasto: R$ {total_gasto:,.2f}
Quantidade:{quantidade:,}
Preço Médio: R$ {preco_medio:,.2f}

Qualquer dúvida é so falar.
Att., César
"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("ctrl", "enter")
```
